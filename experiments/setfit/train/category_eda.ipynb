# 카테고리별로 데이터프레임 분해

```python
from text_sanitizer import *

preprocess = PlainTextPreprocessing()
num_samples = 20000

for category in ["품질", "가격", "서비스", "디자인", "브랜드제품기타"]:
    print(f'\n{category}')
    df_category = df_length[["text", category]]
    category_value_counts = df_category[category].value_counts()
    min_criteria = category_value_counts.min()
    df_category_neg = df_category[df_category[category]==0][:min_criteria]
    df_category_pos = df_category[df_category[category]==1][:min_criteria]

    df_train_neg = df_category_neg.sample(frac=0.9)
    df_train_pos = df_category_pos.sample(frac=0.9)
    df_test_neg = df_category_neg.drop(df_train_neg.index)
    df_test_pos = df_category_pos.drop(df_train_pos.index)

    df_train_neg = df_train_neg[:num_samples//2]
    df_train_pos = df_train_pos[:num_samples//2]

    df_train_downsampled = pd.concat([df_train_pos, df_train_neg])
    df_test_downsampled = pd.concat([df_test_pos, df_test_neg])

    df_train_downsampled["text"] = df_train_downsampled["text"].apply(preprocess)
    df_test_downsampled["text"] = df_test_downsampled["text"].apply(preprocess)

    display(df_train_downsampled)
    display(df_test_downsampled)

    df_train_downsampled.to_csv(f'data/df_{category}_{num_samples}_train.csv')
    df_test_downsampled.to_csv(f'data/df_{category}_{num_samples}_test.csv')
```
